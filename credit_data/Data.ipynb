Cleaning process for Credit Customer datasets from kaggle


```python
import pandas as pd
```

```python
pd.__version__
```

```python
data = pd.read_csv("credit_customers.csv")
```

```python
data.columns
```

```python
data.drop(columns=['checking_status', 'duration', 'credit_history', 'purpose',
       'credit_amount', 'savings_status', 'employment',
       'installment_commitment','num_dependents','other_payment_plans'],inplace=True)
```

```python
data
```

```python
data.columns
```

```python
Capital = []
for i in data.columns:
    Capital.append(i.capitalize())
Capital
data.columns =  Capital
data.columns
```

```python
data["Class"] = data["Class"].apply(lambda x:1 if x=="good" else 0)
```

```python
data["Existing_credits"].unique()
```

```python
data["Existing_credits"] = data["Existing_credits"].astype(int)
```

```python
data["Other_parties"].unique()
```

```python
data.to_csv("cleaned_data.csv", index=False)
```
