```python
# 2. Select specific columns
# Q2: Select only the columns:

data[['name', 'age', 'group']]
```

```python
# Filter rows
# Q3: Select all students who are in group A.

data[data.group == 'group A']
```

```python
# 4. Conditional filtering
# Q4: Select all students with age > 40.
data.loc[(data.age>40), ['name', 'age']]
```

```python
# 5. Multiple conditions
# Q5: Select all Female students in group B.

data[(data.gender == 'Female' ) & (data.group == 'group B')]
```

```python
data.loc[data.session == 'MORNING' ,  ['name', 'session', 'scholarship']]
```

```python
#Select all students who:
# are Male
# AND their session is AFTERNOON or AFT

data[(data.gender == 'Male') & (data['session'].isin(['AFTERNOON', 'AFT']))]
```

```python
##Filter using .isin()
# Select all students whose group is in:
# ['group A', 'group C']

# Assuming 'name' should be checked for not-null values
data.loc[data['group'].isin(['group A', 'group C']), ['group' , 'name' ]]
```
